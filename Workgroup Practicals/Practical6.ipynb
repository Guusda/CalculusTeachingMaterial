## Integration techniques

In this workgroup we will learn how to solve more difficult integrals using two pawerful techniques: Integration by parts and integration by substitution.

## Integration by parts
The integration by parts technique is used to take integrals of pruducts. It is a direct consequence of the product rule of differentials:
$$
\text{d} \left(f(x) g(x) \right)= g(x) \text{d} f(x) + f(x) \text{d} g(x)
$$
If we take the integral of both sides we obtain:
$$
\int_a^b g(x) \text{d} f(x) + \int_a^bf(x) \text{d} g(x) = \int_a^b \text{d} \left(f(x) g(x) \right)
$$
This is often re-written as follows:
$$
\int_a^b g(x) \frac{\text{d} f(x)}{\text{d} x} \text{d}x = (f(b)g(b) - f(a)g(a)) - \int_a^b f(x) \frac{\text{d} g(x)}{\text{d} x} \text{d}x
$$

## Solved exercise 1
Let's try to use this rule for computing the integral
$$
\int_0^1 e^x x \text{d}x
$$

### Solution
We know that 
$$
\text{d}e^x = e^x \text{d} x
$$
Therefore:
$$
\int_0^1 e^x x \text{d}x = \int_0^1 x \text{d} e^x
$$
We can now use the integration by parts formula:
$$
\int_0^1 x \text{d}e^x + \int_0^1 e^x \text{d}x = 1 e^1 - 0 e^0 = e
$$
Therefore:
$$
\int_0^1 e^x x \text{d}x = \int_0^1 x \text{d}e^x  = e - \int_0^1 e^x \text{d}x
$$
We now just need to solve an easier integral:
$$
\int_0^1 e^x \text{d}x = e^1 - e^0 = e - 1
$$
therefore we can conclude that:
$$
\int_0^1 e^x x \text{d}x = e - (e-1) = 1
$$

## Integration by change of variable 
TIntegration by change of variable, also known as integration by substitution is another useful technique. it is a simple consequence of the chain rule of differentials:
$$
\text{d} y(x) = \frac{\text{d}y(x)}{\text{d} x} \text{d} x
$$

Consider an integral of the following form:

$$
\int_a^b f(y(x)) \frac{\text{d}y(x)}{\text{d} x} \text{d} x
$$

We can use the change of variable formula to trasform it in the potentially simpler integral:

$$
\int_a^b f(y(x)) \text{d} y(x)
$$

## Solved exercise 2
Let's use this formula to solve the following integral:
$$
\int_0^1 x e^{x^2} \text{d} x
$$
We can introduce the new variable $y(x) = x^2$. Its differential is 
$$
\text{d}y(x) = \text{d} x^2 = 2 x \text{d} x
$$
We can now use this formula to transform the original integral:
$$
\int_0^1 x e^{x^2} \text{d} x = \int_0^1 \frac{1}{2} e^{x^2} \left(2 x \text{d} x \right) = \frac{1}{2} \int_0^1 e^{x^2} d x^2  = \frac{1}{2} \int_0^1 e^{y(x)} d y(x)
$$
this final integral is easy to solve since 
$$
e^{y(x)} \text{d} y(x) = \text{d} e^{y(x)}~.
$$
Hence:
$$
\frac{1}{2} \int_0^1 e^{y(x)} d y(x) = \frac{1}{2} \int_0^1  d e^{y(x)} = \frac{1}{2} (e^{y(1)} - e^{y(0)} ) = \frac{1}{2}(e^{1^2} - 1) = \frac{1}{2}(e - 1)~.
$$
